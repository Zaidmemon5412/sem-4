```python
import pandas as pd

df = pd.DataFrame({
    "city": ["NY", "NY", "LA", "LA", "LA"],
    "category": ["A", "B", "A", "A", "B"],
    "sales": [100, 200, 50, 70, 30],
    "units": [1, 3, 1, 2, 2],
    "date": pd.to_datetime(["2026-01-01","2026-01-02","2026-01-01","2026-01-03","2026-01-04"])
})
df
```

```python
df.groupby('city')['sales'].count()
```

```python
df.groupby(['city','category'],as_index=False)['sales'].count()
```

```python
df.groupby("city")[["sales", "units"]].sum()
```

```python
df.groupby('city').agg({'sales':['sum','max'],'units':['sum','max']})
```

```python
out = df.groupby("city").agg(
    total_sales=("sales", "sum"),
    avg_sales=("sales", "mean"),
    total_units=("units", "sum"),
    n_orders=("sales", "size")
).reset_index()
out
```

```python
df.groupby('city').filter(lambda x: x['sales'].sum()>100)
```

```python
df["city_avg_sales"] = df.groupby("city")["sales"].transform("mean")
```

```python
df.groupby('city').filter(lambda x: x['sales'].sum() > 150)
```

```python
df
```

```python
df["city_sales_share"] = df["sales"] / df.groupby("city")["sales"].transform("sum")
df
```
