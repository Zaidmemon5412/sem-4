```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame([[1,np.nan,3,4],[2,4,np.nan,8],[np.nan,4,5,9],
                 [np.nan,1,np.nan,3],[np.nan,np.nan,np.nan,np.nan],[1,1,1,1]],columns=["A","B","C","D"])
```

```python
df
```

```python
df.isna()
```

```python
df.dropna(how="any")
```

```python
df.dropna(how="any",axis=0) # row
```

```python
df.dropna(how="any",axis=1) #col
```

```python
df.dropna(how="all")
```

```python
df.dropna(how="all",subset=["B","D"]) # any means or, all means and
```
