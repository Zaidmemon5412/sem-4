```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

1. Accuracy = (TN + TP) / (TN + TP + FN + FP)
2. Error = 1 - Accuracy = (FN + FP) / (TN + TP + FN + FP)
3. Precision = TP / (TP + FP)
4. Recall(Sensitivity) = TP / (FP + FN)
5. Specificity = TN / (TN + FP)
6. F1 Score = (2 x P x R) / (P + R)

    predicted n(cat) predicted p(dog)
actual n(cat)      TN=1               FP=2

actual p(dog)      FN=2               TP=2

```python
TN=1
TP=2
FN=2
FP=2
print("Accuracy:",(TN + TP) / (TN + TP + FN + FP))
print("Error:",(FN + FP) / (TN + TP + FN + FP))
P=TP / (TP + FP)
R=TP / (FP + FN)
print("Precision:",P)
print("Recall:",R)
print("Specificity:",TN / (TN + FP))
print("F1 score:",(2 * P * R) / (P + R))
```

```python
df=pd.read_csv("diabetes.csv")
df.columns
```

Underweight: Below 18.5
Healthy Weight: 18.5 – 24.9
Overweight: 25.0 – 29.9
Obesity: 30.0>

```python
df["BMI_CAT"]=pd.cut(df.BMI,bins=[0,18.5,24.9,29.9,100],labels=['Underweight','Healthy Weight','Overweight','Obesity'])
```

```python
df.head()
```

```python
df[df.BMI_CAT.isna()==True]
```

```python
df=df.dropna()
```

```python
sns.histplot(data=df,x='BMI')
```

```python
x=df.drop(["BMI_CAT","Outcome"],axis=1)
y=df['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
for i in range(21,45):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    print(accuracy_score(y_test,y_pred))
```
