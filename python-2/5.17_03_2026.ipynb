```python
import pandas as pd

customers = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "name": ["A", "B", "C"]
})

orders = pd.DataFrame({
    "order_id": [101, 102, 103],
    "customer_id": [1, 1, 3],
    "amount": [50, 20, 99]
})

out = pd.merge(customers, orders, on="customer_id", how="left")
out
```

```python
df1=pd.DataFrame({"name":["vha","psp","vby","mss","dps"],
                 "sub":["python","fsd","coa","dm","toc"],
                 "in":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"marks":[9,8,10,8,8]})
```

```python
df1.merge(df2,left_on="in",right_on="id")
```

```python
df1=pd.DataFrame({"name":["vha","psp","vby","mss","dps"],
                 "sub":["python","fsd","coa","dm","toc"],
                 "id":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"marks":[9,8,10,8,8],"fac":["vha","nas","vby","mss","dps"]})
```

```python
df1.merge(df2,left_on="name",right_on="fac",how="outer",suffixes=("_B1_SEM3","_B1_SEM4"))
```
