```python
q1 = auto.horsepower.quantile(0.25)
q1
```

```python
q3 = auto.horsepower.quantile(0.75)
q3
```

```python
auto.horsepower.describe()
```

```python
iqr = q3-q1
iqr #threshold frequency = 1.5 , upperlimit = q3 * 1.5 iqr
```

```python
hl = q3 + 1.5*iqr
ll = q1 - 1.5*iqr
hl,ll
```

```python
wo = auto[(auto["horsepower"] > ll) & (auto["horsepower"] < hl)]
wo
```

```python
sms.boxenplot(x="horsepower",data=wo,color='green')
```
